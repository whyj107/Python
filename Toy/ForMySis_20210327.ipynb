```python
%matplotlib inline
import pandas as pd                # 데이터를 저장하고 처리하는 패키지
import matplotlib.pyplot as plt    # 그래프를 그리는 패키지

fileName = './test0.xlsx'
sheetName = '데이터_1_1'
sv1 = 0.00489258
sv2 = 0.80008942

# 데이터 불러오기
df = pd.read_excel(fileName, sheet_name=sheetName)
```

```python
# 데이터 가공
import numpy as np
# 특정 값이 존재하는 index 값 가져오기
l = df.index[(df['전압(V)'] == sv1) | (df['전압(V)'] == sv2)].tolist()
df2 = df.copy()
# 거꾸로 추가하면 index값은 변화하지 않음
for i in reversed([x+1 for x in l]):
    df2 = pd.DataFrame(np.insert(df2.values, i, values = [np.nan, np.nan, np.nan], axis=0))
# 데이터 미리 보기
plt.plot(df2[1], df2[2])
plt.show()    
```

```python
# 엑셀 저장하는 부분
with pd.ExcelWriter("./output.xlsx") as writer:
    df2.to_excel(writer,sheet_name="Result",index=False)
```
